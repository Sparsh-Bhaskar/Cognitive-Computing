Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find i. Sum of all elements ii. Sum of all elements row-wise iii. Sum of all elements column-wise

```python
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print("sum of all elements: ",np.sum(gfg))
print("sum of elements row wise: \n",np.sum(gfg,axis=1))
print("sum of elements column wise: ",np.sum(gfg,axis=0))
```

Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find i. Sorted array ii. Indices of sorted array iii. 4 smallest elements iv. 5 largest elements

```python
arr=np.array([10,52,62,16,16,54,453])
print("sorted array: ",np.sort(arr))
print("indices of sorted array: ",np.argsort(arr))
print("smallest elements: ",np.min(arr))
print("largest elements: ",np.max(arr))
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find i. Integer elements only ii. Float elements only

```python
arr = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print("integer elements: ")
print([x for x in arr if isinstance(x, int)])
print(arr[arr.astype(int) == arr])
print("float elements: ")
print([x for x in arr if isinstance(x, float)])
print(arr[arr.astype(float) ==arr])
```

Q.3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting. a) Generate your unique sales dataset:  Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65

66 = 131 → sales = [131, 181, 231, 281, 331].)  Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200].

```python
name=str("SB")
l=len(name)
X=0
for i in range(l):
  X+=ord(name[i])

print(X)
sales=np.array([X,X+50,X+100,X+150,X+200])
print("sales array: ",sales)
```

(b) Compute your personalized tax rate as ((X % 5) + 5) / 100.  Use broadcasting to apply this tax rate to each sales value.

```python
tax_rate=((X%5)+5)/100
print("tax rate is ",tax_rate)
print("after apllying tax_rate to each sales value: ")
res=sales+sales*tax_rate
print(res)
```

(c) Adjust sales based on discount:  If sales < X+100, apply a 5% discount.  If sales >= X+100, apply a 10% discount.

```python
print("original rates")
print(sales)
print("after tax")
print(res)
sd=res
for i in range(len(sd)):
  if sd[i]==X+100:
    sd[i]-=sd[i]*0.1
print("discounted rates")
print(sd)
```

(d) Expand sales data for multiple weeks:  Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.  Increase sales by 2% per week using element-wise broadcasting

```python
sales_matrix=np.tile(sd,(3,1))
print(sales_matrix,"\n")
weeks=np.arange(3).reshape(3,1)
sales_matrix*=(1+0.02*weeks)
print(sales_matrix)
```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:  Y = x2  Y = sin(x)  Y = ex  Y = log(|x| + 1) Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
y=x**2
plt.figure()
plt.plot(x,y)
plt.title("plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

y=np.sin(x)
plt.figure()
plt.plot(x,y)
plt.title("plot of y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

y=np.exp(x)
plt.figure()
plt.plot(x,y)
plt.title("plot of y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

y= np.log(abs(x)+1)
plt.figure()
plt.plot(x,y)
plt.title("plot of y = log(|x|+1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```
