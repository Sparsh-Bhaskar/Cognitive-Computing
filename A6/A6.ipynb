## Q1: Mathematical Function Plotting
**Task:**  
Create a program that:
1. Takes user input for a value M
2. Generates x values from -10 to 10 using `np.linspace()`
3. Computes two functions:
   - y = M·x²
   - y = M·sin(x)
4. Plots both functions in one figure with:
   - Different colors/line styles
   - Legend, grid, and title

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

M = float(input("Please enter a value for M: "))

x = np.linspace(-10, 10, 400)

y_quadratic = M * x**2
y_sine = M * np.sin(x)

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

plt.plot(x, y_quadratic, color='blue', linestyle='-', linewidth=2, label=f'{M}·x²')
plt.plot(x, y_sine, color='red', linestyle='--', linewidth=2, label=f'{M}·sin(x)')

plt.title(f'Mathematical Functions with M = {M}', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Q2: Subject Scores Bar Plot
**Task:**  
1. Create a dataset of 5 subjects with scores
2. Convert to Pandas DataFrame
3. Visualize using Seaborn bar plot with:
   - Distinct colors per bar
   - Value annotations on bars
   - Title, axis labels, and grid

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'Computer Science']
marks = [85, 78, 92, 88, 95]

df = pd.DataFrame({
    'Subject': subjects,
    'Marks': marks
})

plt.figure(figsize=(8, 5))

sns.barplot(x='Subject', y='Marks', data=df, palette='pastel')

for index, row in df.iterrows():
    plt.text(index, row['Marks']+1, str(row['Marks']),
             ha='center', fontsize=12)

plt.title('Subject Marks Comparison', fontsize=14)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

## Q3: Random Data Visualization
**Task:**  
1. Set NumPy random seed using your roll number
2. Generate 50 values using `np.random.randn()`
3. Create 2×2 subplots showing:
   - Line plot of cumulative sum
   - Scatter plot with random noise
4. Customize with:
   - Titles
   - Axis labels
   - Grids

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

data = np.random.randn(50)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
cumulative_sum = np.cumsum(data)
plt.plot(cumulative_sum, color='blue')
plt.title("Cumulative Sum of Values")
plt.xlabel("Index")
plt.ylabel("Total")
plt.grid(True)

plt.subplot(2, 2, 2)
x_values = np.arange(50)
noise = np.random.randn(50) * 2
plt.scatter(x_values, data + noise, color='red', alpha=0.6)
plt.title("Scatter Plot with Noise")
plt.xlabel("Position")
plt.ylabel("Value")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Q4: Sales Data Analysis
**Dataset:**  
[company_sales_data.csv](https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv)

**Tasks:**  
Using Seaborn:
1. Line plot of total profit by month
2. Multiline plot of all product sales
3. Bar chart of all features/attributes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)

sns.set_style("whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(x='month_number', y='total_profit', data=sales_data, marker='o')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Profit in USD')
plt.show()

plt.figure(figsize=(10, 5))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    sns.lineplot(x='month_number', y=product, data=sales_data, label=product, marker='o')
plt.title('Sales of All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.show()

numeric_data = sales_data.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12, 6))
numeric_data.plot(kind='bar', figsize=(12, 6))
plt.title('Sales Data for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(range(len(sales_data)), sales_data['month_number'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
